```python
import os
import pandas as pd

def read_all_csv_files_in_directory(directory):
    # 创建一个空的列表来存储所有的 DataFrame
    dataframes = []

    # 使用 os.walk 来遍历目录及其子目录
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.csv'):
                # 构建文件的完整路径
                file_path = os.path.join(root, file)
                print(file[:-4])
                # 读取 CSV 文件到 DataFrame
                df = pd.read_csv(file_path)

                # df.to_sql(file[:-4], engine, if_exists='replace', index=False)

                # 将 DataFrame 添加到列表中
                dataframes.append(df)
    
    # 返回包含所有 DataFrame 的列表
    return dataframes

# 示例用法
directory_path = '../Data'  
all_dataframes = read_all_csv_files_in_directory(directory_path)
```

```python
import psycopg2
import csv

# 连接到 PostgreSQL 数据库
conn = psycopg2.connect(
    dbname="postgres",
    user="tanghao",
    password="",
    host="localhost",
    port="5432"
)

# 创建游标
cur = conn.cursor()

# 创建表的 SQL 语句
create_table_sql = """
CREATE TABLE employees (
    id SERIAL PRIMARY KEY,
    name VARCHAR(100),
    position VARCHAR(50),
    salary NUMERIC(10, 2),
    hire_date DATE
);
"""

# 执行创建表的 SQL 语句
cur.execute(create_table_sql)

# 提交事务
conn.commit()

# 导入 CSV 数据的函数
def import_csv_to_db(csv_file_path):
    with open(csv_file_path, 'r') as f:
        reader = csv.reader(f)
        next(reader)  # 跳过表头
        for row in reader:
            cur.execute(
                "INSERT INTO employees (id, name, position, salary, hire_date) VALUES (%s, %s, %s, %s, %s)",
                row
            )

# 调用导入 CSV 数据的函数
import_csv_to_db('/absolute/path/to/your/file.csv')

# 提交事务
conn.commit()

# 查询数据的 SQL 语句
select_all_sql = "SELECT * FROM employees;"

# 执行查询的 SQL 语句
cur.execute(select_all_sql)

# 获取查询结果
rows = cur.fetchall()

# 打印查询结果
for row in rows:
    print(row)

# 关闭游标和连接
cur.close()
conn.close()
```
